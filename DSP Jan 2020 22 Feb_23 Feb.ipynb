```python
# Create a new column called Zone
# based on Location, determine the Zone
# If Location == New Delhi ---> N
# If Location == Chennai ---> S
# If Location == Kol ---> E
# If Location == Mum ---> W

stores["Zone"] = np.where(stores.Location == "New Delhi","N",np.where(stores.Location == "Chennai","S",np.where(stores.Location == "Mumbai","W",np.where(stores.Location == "Kolkata","E",""))))
```

```python
stores
```

```python
# Binning
#Derive a variable "Performance" from TotalSales

# 	cat			conditions
#    ------			----------
# 	low			 < 100
# 	med			100 - 200
# 	high			200 - 300
# 	vhigh			> 300

ts = stores.TotalSales
stores["Performance"] = np.where((ts < 100), "Low",
	 np.where((ts >= 100) & (ts < 200),"med",
                 np.where((ts >= 200) & (ts < 300),"high",
                          np.where((ts >= 300),"vhigh",""))))
```

```python
# Aggregations/group by
# calculation on continuous var based on a categorical variable

# SELECT StoreType, sum(OperatingCost) as SumOfOperatingCost FROM stores GROUP BY StoreType


# Step 1 : The groups
# df.groupby()
t = stores.groupby(by = ["StoreType"])

# Step 2 : Performing aggregations/calculations on the groups
# sum of OperatingCost

t.OperatingCost.sum()
t["OperatingCost"].sum()

# DataFrame
t[["OperatingCost"]].sum()
```

```python
# Calculate Avg total sales per Location

# Step 1
t = stores.groupby(by = ["Location"])

# Step 2 : calcualte TotalSales.mean()
res1 = t[["TotalSales"]].mean()
res2 = res1.reset_index()
display(res1)
display(res2)
```

```python
res1.reset_index(drop=True)
```

```python
# Group by Location and StoreType, find Average Operating Cost

# Step 1:
t = stores.groupby(by = ["Location","StoreType"])
# Step 2:
res1 = t[["OperatingCost"]].mean().add_prefix("Avg_Of_")
display(res1)
res2 = res1.reset_index()
display(res2)
```

```python
# Create a table tat aggregates Location and calculate Sum Of OperatingCost and TotalSales

# Step 1
t = stores.groupby(by="Location")
t

# Aggregation
res1 = t[["OperatingCost","TotalSales"]].sum().add_prefix("Sum_of_")
display(res1)
res2 = res1.reset_index()
display(res2)

```

```python
# Create a table tat aggregates Location and calculate Sum Of OperatingCost and Mean of TotalSalels

# Step 1
t = stores.groupby(by="Location")
t

# Method 1
# a = calc -> Sum of OC
# b = calc -> mean TS
# c = pd.concat([a,b],axis=1)

a = t[["OperatingCost"]].sum().add_prefix("Sum_Of_")
b = t[["TotalSales"]].mean().add_prefix("Mean_Of_")

res1 = pd.concat([a,b],axis=1) # row concat
display(res1)
res2 = res1.reset_index()
display(res2)
```

```python
# Method 2
# use .agg()

# Step1 : 
t = stores.groupby(by="Location")

# Step2
res1 = t[["OperatingCost","TotalSales"]].agg({"OperatingCost":"sum","TotalSales":"mean"})
res1 = res1.rename(columns={"OperatingCost":"Sum_Of_OperatingCost","TotalSales":"Mean_Of_TotalSales"})
res1
res2 = res1.reset_index()
```

```python
# Task : calculate the freq table of StoreType

stores.groupby("StoreType")[["StoreType"]].count().add_prefix("Count_")


a = stores.groupby("StoreType")[["TotalSales"]].count().add_prefix("Count_")
b = stores.groupby("StoreType")[["TotalSales"]].min().add_prefix("Min_")
c = stores.groupby("StoreType")[["TotalSales"]].max().add_prefix("Max_")
pd.concat([a,b,c],axis=1)
```
