```python
def impute_missing(x):
    if   == "float64" or x.dtype == "int64" or x.dtype == "int32":
        return x.fillna(x.median())
    elif x.dtype == "object":
        mode_x = x.value_counts().index[0]
        return x.fillna(mode_x)
    else:
        return x
    
```

```python
# Continous
#   count
#   pct_miss
#   sum
#   mean
#   min 0%
#   1%
#   5%
#   10%
#   Q1 - 25%
#   median - 50%
#   Q3 - 75%
#   90%
#   95%
#   99%
#   max - 100%

stores.AcqCostPercust.quantile([0,1])
# takes 0 to 1 as input
# 0 -> 0%
# 1 -> 100%
acq_quant = stores.AcqCostPercust.quantile([0,0.01,0.05,0.1,0.25,0.5,0.75,0.9,0.95,0.99,1])

acq_quant.index = ["min", "p01", "p05", "p10", "p25", "median", "p75", "p90", "p95", "p99", "max"]

acq_quant
```

```python
def numerical_summary(x):
    n = len(x)
    pct_miss = (n - x.count())/n * 100
    total = round(x.sum(),2)
    avg = round(x.mean(),2)
    q = x.quantile([0,0.01,0.05,0.1,0.25,0.5,0.75,0.9,0.95,0.99,1])
    q.index = ["min", "p01", "p05", "p10", "p25", "median", "p75", "p90", "p95", "p99", "max"]
    res =  pd.Series([n,pct_miss,total,avg], index = ["length","pct_missing","total","average"])
    
    return pd.concat([res,q],axis=0)
```

```python
numerical_summary(stores.AcqCostPercust)
```

```python
stores_cont.apply(numerical_summary)
```

```python
stores_cont.apply(numerical_summary).T
```

```python
# Categorical
#    count
#    pct_missing
#    mode
#    top_3_cat

def categorical_summary(x):
    n = len(x)
    pct_miss = round((n - x.count())/n * 100,2)
    mode_x = x.value_counts().index[0]
    top_3 = list(x.value_counts().index[:3])
    top_3 = [str(x) for x in top_3]
    top_3_str = ", ".join(top_3)
    
    return pd.Series([n,pct_miss,mode_x,top_3_str], index = ["length","pct_missing","mode","top_3"])
```
